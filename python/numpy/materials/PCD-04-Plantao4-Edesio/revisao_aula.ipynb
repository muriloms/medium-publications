## Comandos Básicos

```python
import numpy as np
```

```python
# Criar array 1d
a = [1,3,5,7,9,10]
b1d = np.array(a)
b1d
```

```python
# Criar array 2d
a = [[1,3,5],[7,9,10]]
b2d = np.array(a)
b2d
```

```python
# retorna array de zeros
np.zeros(shape=(5,2))
```

```python
# retorna array de uns
np.ones(shape=(5,2))
```

```python
# retorna um array com dados espaçados
np.arange(start=1, stop=10, step=2)
```

```python
# retorna array espaçado em um intervalo fixo
np.linspace(start=1, stop=10, num=100)
```
