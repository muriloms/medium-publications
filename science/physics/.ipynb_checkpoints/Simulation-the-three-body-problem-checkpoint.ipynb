```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definindo as constantes e condies iniciais
t0, tf = 0., 500.  # Tempo inicial e final
h = 0.1  # Incrementos de tempo
n = 12  # Nmero de componentes para y e funo fora
r1 = np.array([10., 0.]);   v1 = np.array([0., -0.1]); m1 = 1.  # Corpo 1
r2 = np.array([-10., 0.]);  v2 = np.array([0., 0.1]); m2 = 1.  # Corpo 2
r3 = np.array([0., 10.]);   v3 = np.array([-0.1, 0.]); m3 = 1.  # Corpo 3
y0 = np.concatenate((r1, r2, r3, v1, v2, v3))  # Condies iniciais

def f(y, t):
    r1, r2, r3 = y[:2], y[2:4], y[4:6]
    v1, v2, v3 = y[6:8], y[8:10], y[10:12]
    
    # Distncias relativas
    r12 = np.linalg.norm(r2 - r1)
    r13 = np.linalg.norm(r3 - r1)
    r23 = np.linalg.norm(r3 - r2)
    
    # Aceleraes
    a1 = m2 * (r2 - r1) / r12**3 + m3 * (r3 - r1) / r13**3
    a2 = m1 * (r1 - r2) / r12**3 + m3 * (r3 - r2) / r23**3
    a3 = m1 * (r1 - r3) / r13**3 + m2 * (r2 - r3) / r23**3
    
    dydt = np.concatenate((v1, v2, v3, a1, a2, a3))
    return dydt

# Criando o vetor de tempo
t = np.arange(t0, tf, h)

# Resolvendo as EDOs
solution = odeint(f, y0, t)

# Extraindo as solues para as posies dos corpos
r1_sol = solution[:, :2]
r2_sol = solution[:, 2:4]
r3_sol = solution[:, 4:6]

# Plotando as trajetrias dos trs corpos
plt.figure(figsize=(10, 10))
plt.plot(r1_sol[:, 0], r1_sol[:, 1], label='Corpo 1')
plt.plot(r2_sol[:, 0], r2_sol[:, 1], label='Corpo 2')
plt.plot(r3_sol[:, 0], r3_sol[:, 1], label='Corpo 3')

plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.legend()
plt.title('Trajetrias dos trs corpos')
plt.grid()
plt.show()
```
